```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_excel('Datasets/Your sugary status (Responses).xlsx')
df.drop('Timestamp',axis=1,inplace = True)
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['How often you think about your sugar intake? '],data=df)
```

```python
df.groupby(df['How often you think about your sugar intake? ']).size()
```

```python
# How many responders knows their diabetic status

sns.countplot(x=df['Do you know about your current diabetic status? (Status refers to if you know you are diabetic or not)']
              ,data=df)
```

```python
# Which gender has more awareness about their diabetic status

sns.countplot(x=df['Do you know about your current diabetic status? (Status refers to if you know you are diabetic or not)'],
             hue='Gender', data=df)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['Age group'], data=df)
plt.xlabel('Ages')
plt.ylabel('Ages count')
```

```python
# Responses against age group
df.groupby(df['Age group']).size()
```
